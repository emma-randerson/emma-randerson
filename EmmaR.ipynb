Create a table containing the total distance travelled by each carrier (you can get this from the flights and airlines data).  

```python
carrier_dist = flights.merge(airlines, how='left', on='carrier')
carrier_dist.head(5)
```

```python
carrier_dist = carrier_dist[['carrier','name','distance']]
```

```python
# To check bulk sales
carrier_dist = carrier_dist.groupby(
   ['carrier','name']
).agg(
    {
         'distance': "sum",
    }
).round(decimals=2)
carrier_dist = carrier_dist.reset_index()
```

```python
carrier_dist
```

```python
from io import StringIO  
csv_buffer = StringIO()
carrier_dist.to_csv(csv_buffer, index=False)
s3_resource = boto3.resource('s3')
s3_resource.Object('kilimanjaro-prod-datalake','newstarter/datascience/emmaranderson/carrier_dist.csv').put(Body=csv_buffer.getvalue())
```
